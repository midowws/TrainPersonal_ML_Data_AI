```python
import pandas as pd
import joblib
import sklearn
from sklearn.ensemble import RandomForestClassifier

model_load = joblib.load('model_random_forest.pkl')

nama_bunga =['Setosa', 'Versicolor', 'Virginica']

sl = float(input("Masukkan panjang sepal: "))
sw = float(input("Masukkan lebar sepal: "))
pl = float(input("Masukkan panjang petal: "))
pw = float(input("Masukkan lebar petal: "))

data_siap_prediksi = [[sl, sw, pl, pw]]

hasil = model_load.predict(data_siap_prediksi)
print(f"Hasil Prediksi AI: {nama_bunga[hasil[0]]}")

```
